# 3.Data Wrangling

```python
# Extract the 3-letter currency code from the beginning of each price string and save in new column 'Currency'
df_boat['Currency'] = df_boat['Price'].str[:3]
```

```python
# Currency frequency values
df_boat['Currency'].value_counts(dropna = False)
```

```python
#Removes everything except digits
df_boat['Price'] = df_boat['Price'].str.replace(r'\D', '', regex=True)
```

```python
#Check the df
df_boat.head()
```

```python
# change data type of price to interger so that we can convert all curriencies to similar
df_boat['Price'] = df_boat['Price'].astype('int')
```

```python
# converting currency into euro
def europrice(Price,Currency):
    if Currency=='CHF':
        return Price * 1.02
    elif Currency=='DKK':
        return Price * 0.13
    elif Currency=='Â£':
        return Price * 1.17
    else:
        return Price
df_boat['EUR Price'] = df_boat.apply(lambda x: europrice(x.Price, x.Currency), axis=1)
```

```python
#Check output
df_boat.head()
```

```python
# Extract country,city and standing point from the Location 
df_boat[['Country', 'City', 'Standing_point']] = df_boat['Location'].str.extract(r'([^Â»]+)Â»([^Â»]+)Â»(.+)', expand=True)
```

```python
# check the df
df_boat.head()
```

```python
# extract Country and City from Location
df_boat[['Country1','City1','Standing_point1']] = df_boat['Location'].str.split('Â»', n=2, expand=True)
```

```python
# check the dataframe
df_boat.head()
```

```python
# Drop the redundant columns
df_boat = df_boat.drop(['Price','Currency','Location','Standing_point','Country','City','Standing_point1'], axis = 1)
```

```python
#Check the df
df_boat.head()
```

# 4.Data Cleaning

```python
# Check for Missing values
df_boat.isnull().sum()
```

```python
#Filling length and width with mean values
df_boat['Length'] = df_boat['Length'].fillna(df_boat['Length'].median())
df_boat['Width'] = df_boat['Width'].fillna(df_boat['Width'].median())
```

```python
#Filling country1 and city1 with unknown values
df_boat['Country1'] = df_boat['Country1'].fillna('Unknown')
df_boat['City1'] = df_boat['City1'].fillna('Unknown')
```

```python
# check the shape of df
df_boat.shape
```

```python
# Statistical analysis
df_boat.describe()
```

```python
# Drop the column 'year built' with 0
df_boat_clean = df_boat.loc[df_boat['Year Built'] != 0]
```

```python
# check for duplicates
df_boat_clean_dups = df_boat_clean[df_boat_clean.duplicated()]
```

```python
# check for mixed data type
for col in df_boat_clean.columns.tolist():
  weird = (df_boat_clean[[col]].applymap(type) != df_boat_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_boat_clean[weird]) > 0:
    print (col)
```

```python
#INspect the data to identigy rows with mixed data
print(df_boat_clean[['Manufacturer', 'Type', 'Material']].dtypes)
```

```python
#Convert columns to a single type 
df_boat_clean.loc[:, 'Manufacturer'] = df_boat_clean['Manufacturer'].astype(str)
df_boat_clean.loc[:, 'Type'] = df_boat_clean['Type'].astype(str)
df_boat_clean.loc[:, 'Material'] = df_boat_clean['Material'].astype(str)
```

```python
#Decode the string to correct the characters to resolve the character encoding issues, where special characters are not rendered properly.
df_boat_clean.loc[:, 'Manufacturer'] = df_boat_clean['Manufacturer'].apply(
    lambda x: x.encode('latin1', errors='ignore').decode('utf-8', errors='ignore') if isinstance(x, str) else x
)
```

```python
# drop the other missing values
df_boat_clean = df_boat_clean.dropna()
```

```python
# check for the shape of the cleaned dataframe
df_boat_clean.shape
```

```python
# check for mixed data type
for col in df_boat_clean.columns.tolist():
  weird = (df_boat_clean[[col]].applymap(type) != df_boat_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_boat_clean[weird]) > 0:
    print (col)
```
